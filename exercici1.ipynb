```python
import nltk
# Descarreguem l'Europarl Corpus
nltk.download('europarl_raw')
import nltk.corpus.europarl_raw as europarl
# Obtenim la llista de documents a europarl per l'Anglès
europarl.english.fileids()
# Obtenim el text del document 'ep-00-01-17.en'
raw_text = europarl.english.raw('ep-00-01-17.en')
```

Tokenitzeu el document 'ep-00-01-17.en' amb els tokenitzadors descrits en
aquesta sessió.


```python
# Tokenització amb NLTK 
tokens_nltk = [nltk.word_tokenize(s, language='english') for s in
 nltk.sent_tokenize(raw_text, language='english')]

tokens_nltk
```

```python
# Tokenització ambspaCy
# Model Anglès: en_core_web_sm. No cal baixar el model en aquest cas, el carreguem directaament.
import spacy
nlp = spacy.load('en_core_web_sm')

doc = nlp(raw_text)

tokens_spacy =  [(token.text, token.is_stop) 
                 for s in doc.sents
                 for token in s]

tokens_spacy
```

```python
# Tokenització ambTextServer (FreeLing)

from textserver import TextServer

ts = TextServer('usuari', 'passwd', 'tokenizer')
tokens_textserver = ts.tokenizer(raw_text)

tokens_textserver
```
