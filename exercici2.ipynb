```python
import nltk
from nltk.collocations import BigramCollocationFinder
from nltk import word_tokenize
import nltk.corpus.europarl_raw as europarl
nltk.download('punkt')
```

Calculeu els 10 2-grams més freqüents (ordenats) per a l'arxiu europarl.english.raw('ep-00-01-17.en').

```python
# 1. Obtenim el text cru del document 'ep-00-01-17.en' en anglès
raw_text = europarl.english.raw('ep-00-01-17.en')

# 2. Tokenitzem el text per separar-lo en paraules 
tokens = word_tokenize(raw_text)

# 3. Creem el model de bigrames a partir dels tokens 
finder = BigramCollocationFinder.from_words(tokens)

# 4. Extraiem els 10 més freqüents
# finder.ngram_fd conté la distribució de freqüències (FreqDist)
top_10_bigrams = finder.ngram_fd.most_common(10)

print("Els 10 2-grames més freqüents són:")
for bigram, freq in top_10_bigrams:
    print(f"{bigram}: {freq}")
```

Per cada un dels arxius en Anglès (europarl.english.fileids()), calculeu
el percentatge d'aquests 10 2-grams que també es troben en cadascun
d'aquests arxius.

```python
top_10_set = set([bigram for bigram, freq in top_10_bigrams])

print("Percentatge de coincidència en els documents en anglès:")

# Recorrem tots els arxius en anglès del corpus
for fileid in europarl.english.fileids():
    # 1. Llegim el text de l'arxiu actual
    text_arxiu = europarl.english.raw(fileid)
    
    # 2. Tokenitzem el text
    tokens_arxiu = word_tokenize(text_arxiu)
    
    # 3. Creem els bigrames d'aquest arxiu
    finder_arxiu = BigramCollocationFinder.from_words(tokens_arxiu)
    
    # Obtenim tots els bigrames únics d'aquest arxiu com a conjunt
    bigrames_arxiu = set(finder_arxiu.ngram_fd.keys())
    
    # 4. Calculem quants dels nostres 10 bigrames apareixen aquí
    coincidencies = top_10_set.intersection(bigrames_arxiu)
    
    # 5. Calculem el percentatge
    percentatge = (len(coincidencies) / len(top_10_set)) * 100
    
    print(f" - {fileid}: {percentatge}% dels 2-grames trobats.")
```

Feu el mateix per a l'arxiu 'ep-00-01-18', en aquest cas pels diferents idiomes: danish, dutch, english, finnish, french, german, greek, italian,
portuguese, spanish i swedish.

```python
# Diccionari amb els idiomes demanats i la seva extensió d'arxiu corresponent
idiomes_ext = {
    'danish': 'da',
    'dutch': 'nl',
    'english': 'en',
    'finnish': 'fi',
    'french': 'fr',
    'german': 'de',
    'greek': 'el',
    'italian': 'it',
    'portuguese': 'pt',
    'spanish': 'es',
    'swedish': 'sv'
}

arxiu_base = 'ep-00-01-18'

print(f"Percentatge de coincidència per a l'arxiu {arxiu_base} en diferents idiomes:")

for idioma, ext in idiomes_ext.items():
    # Construïm el nom exacte de l'arxiu, per exemple 'ep-00-01-18.es'
    nom_arxiu = f"{arxiu_base}.{ext}"
    # Això equival a fer europarl.spanish, europarl.french, etc.
    corpus_idioma = getattr(europarl, idioma)
    
    
    # 1. Llegim el text cru
    text_arxiu = corpus_idioma.raw(nom_arxiu)
    
    # 2. Tokenitzem
    tokens_arxiu = word_tokenize(text_arxiu)
    
    # 3. Creem els bigrames i els passem a un conjunt
    finder_arxiu = BigramCollocationFinder.from_words(tokens_arxiu)
    bigrames_arxiu = set(finder_arxiu.ngram_fd.keys())
    
    # 4. Comprovem les coincidències amb el nostre top_10_set (de l'anglès)
    coincidencies = top_10_set.intersection(bigrames_arxiu)
    percentatge = (len(coincidencies) / len(top_10_set)) * 100
    
    print(f" - {idioma.capitalize()}: {percentatge}%")
```

És suficient aquest model per identificar l'idioma Anglès?
- Sí, és molt efectiu. Els 2-grames més freqüents d'un idioma (com ara "of the" o "in the" en anglès) actuen com una signatura estructural gairebé única.
- Aquestes combinacions apareixen de forma constant en qualsevol text anglès independentment del tema, mentre que la seva aparició en textos d'altres idiomes és pràcticament nul·la (0%).

Què passa si fem servir 1-grams? I 5-grams?
- Amb 1-grames (paraules aïllades): El model és menys robust. Tot i que pot funcionar, augmenta el risc de falsos positius perquè paraules molt curtes (com "a", "en", "me") coincideixen freqüentment amb el vocabulari d'altres idiomes.
- Amb 5-grames (seqüències llargues): El model fracassa per l'esparsió de dades (data sparsity). Les combinacions exactes de 5 paraules depenen massa del contingut específic del text (són massa concretes). Per tant, gairebé mai es repetiran en altres documents, impedint que el model reconegui el propi idioma.
